```python
import pandas as pd
```

```python
csvpath = "purchase_data copy.csv"
print(csvpath)
purchase_df = pd.read_csv(csvpath)
purchase_df.head()
```

```python
purchase_df.count()
```

```python
##1. Player Count
# the number of total players
total_player = len(purchase_df["SN"].unique())
total_player_df = pd.DataFrame({"Total Players":[total_player]})
total_player_df
```

```python
##2. Purchasing Analysis (Total)
# the number of unique items
unique_items = purchase_unique_df = purchase_df["Item ID"].unique()
unique_number = len(unique_items)
# the number of purchase
purchase_number = len(purchase_df["Price"])
# the average price
average_price = purchase_df["Price"].mean()
average_price = "${:.2f}".format(average_price)
# total revenue
total_revenue = "${:.2f}".format(purchase_df["Price"].sum())   
# summary frame
summary = {"Number of Unique Items":[unique_number],"Average Price":[average_price],"Number of Purchases":[purchase_number],"Total Revenue":[total_revenue]}
summary_df = pd.DataFrame(summary)
summary_df
```

```python
##3. Gender Demographics
gender_data_df = purchase_df.groupby(["Gender"])
gender_player = gender_data_df["SN"].nunique()
gender_player_df = pd.DataFrame(gender_player)
gender_player_df["Percentage of Players"] = (gender_player_df["SN"]/total_player).map("{:.2%}".format)
gender_player_df = gender_player_df.rename(columns={"SN":"Total Count"})
gender_player_df
```

```python
##4. Purchasing Analysis (Gender)
# Purchase Count
purchase_count = gender_data_df.count()["Purchase ID"]
# Average Purchase Price
average_pruchase_price_unf = gender_data_df["Price"].mean()
average_pruchase_price_f = average_pruchase_price_unf.map("${:.2f}".format)
# Total Purchase Value
gender_total_unf = gender_data_df["Price"].sum()
gender_total_f = gender_total_unf.map("${:.2f}".format)
# Avg Total Purchase per Person
gender_avg_person_unf = gender_total_unf/gender_player_df["Total Count"]
gender_avg_person_f = gender_avg_person_unf.map("${:.2f}".format)
# gender_summary = pd.DataFrame({"Purchase Count":[],"Average Purchase Price":[average_pruchase_price],"Total Purchase Value":[gender_total ],"Avg Total Purchase per Person":[]})
gender_summary = pd.DataFrame({"Purchase Count":purchase_count,"Average Purchase Price":average_pruchase_price_f,"Total Purchase Value":gender_total_f,"Avg Total Purchase per Person":gender_avg_person_f})
gender_summary
```

```python
##5.Age Demographics
bins = [0,10,15,20,25,30,35,40,100]
group_name = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_df["Age Summary"] = pd.cut(purchase_df["Age"],bins,labels=group_name,right = False, include_lowest = True)
age_demo = purchase_df.groupby("Age Summary")["SN"].nunique()
age_demo_df = pd.DataFrame(age_demo)
age_demo_df = age_demo_df.rename(columns={"SN":"Total Count"})
age_demo_df["Percentage of Players"] = (age_demo_df["Total Count"]/total_player).map("{:.2%}".format)
age_demo_df
```

```python
##6.Purchasing Analysis (Age)
# Purchase Count
age_purchase_count = purchase_df.groupby("Age Summary")["Price"].count()
# Average Purchase Price
age_average_price = purchase_df.groupby("Age Summary")["Price"].mean()
# Total Purchase Value
age_total = purchase_df.groupby("Age Summary")["Price"].sum()
# Avg Total Purchase per Person
age_avg_person = (age_total/age_demo_df["Total Count"]).map("${:.2f}".format)
age_summary = pd.DataFrame({"Purchase Count":age_purchase_count,"Average Purchase Price":age_average_price,"Total Purchase Value":age_total,"Avg Total Purchase per Person":age_avg_person})
age_summary["Average Purchase Price"] = age_summary["Average Purchase Price"].map("${:.2f}".format)
age_summary["Total Purchase Value"] = age_summary["Total Purchase Value"].map("${:.2f}".format)
age_summary
```

```python
##7.Top Spenders
spender_df = purchase_df.groupby(["SN"])
# Purchase Count
spender_purchase_count = spender_df.count()["Age"]
# Average Purchase Price
spender_average = spender_df["Price"].mean()
# Total Purchase Value
spender_total = spender_df["Price"].sum()
# Spender Summary
summary_spender = pd.DataFrame({"Purchase Count":spender_purchase_count,"Average Purchase Price":spender_average,"Total Purchase Value":spender_total})
summary_spender_des = summary_spender.sort_values("Total Purchase Value",ascending = False)
summary_spender_des["Average Purchase Price"] = summary_spender_des["Average Purchase Price"].map("${:.2f}".format)
summary_spender_des["Total Purchase Value"] = summary_spender_des["Total Purchase Value"].map("${:.2f}".format)
summary_spender_des.head()
```

```python
##8.Most Popular Items
popular_df = purchase_df.groupby(["Item ID","Item Name"])
# Purchase Count
purchase_count = popular_df.count()["Price"]
# Items Price
items_price = total_value = popular_df["Price"].mean().map("${:.2f}".format)
# Total Purchase Value
total_value = popular_df["Price"].sum()
# Popularity Summary
summary_items = pd.DataFrame({"Purchase Count":purchase_count,"Items Price":items_price,"Total Purchase Value":total_value})
summary_populartity = summary_items.sort_values("Purchase Count",ascending = False)
summary_populartity["Total Purchase Value"] = summary_populartity["Total Purchase Value"].map("${:.2f}".format)
summary_populartity.head()
```

```python
##9.Most Profitable Items
summary_profit = summary_items.sort_values("Total Purchase Value",ascending = False)
summary_profit["Total Purchase Value"] = summary_profit["Total Purchase Value"].map("${:.2f}".format)
summary_profit.head()
```
